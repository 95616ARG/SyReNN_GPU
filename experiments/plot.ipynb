```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
  
colors = ("#4e73b0", "#fdb863", "#b2abd2", "#e66101", "#5e3c99")
label_names = ("Clear-of-Conflict", "Weak Left", "Weak Right", "Strong Left", "Strong Right")

fig, ax = plt.subplots(figsize=(11.2, 8))
plt.figure(frameon=False)
ax.axis('off')
im = img.imread('fhat-all.png')
imagebox = OffsetImage(im, zoom=0.5)
imagebox.image.axes = ax
ab = AnnotationBbox(imagebox, (0.315, 0.5), xycoords='axes fraction',
                    bboxprops={'lw':0}, frameon=False )

ax.add_artist(ab)

legend_elements = [
    Line2D([0], [0],
        # marker='o',
        color=colors[label],
        lw=14,
        label=label_names[label])
    for label in [0,1,2,3,4] ]

ax.legend(handles=legend_elements, loc="upper right", prop={'size': 24})

fig.tight_layout()

plt.show()
fig.savefig(f"fhat-all-intro.png" , dpi=200)
```
